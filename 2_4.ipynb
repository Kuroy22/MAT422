MLE for random samples

```python
import numpy as np
import pandas as pd
#creates independent variable
x=np.linspace(-20, 40 , 100)
#generates a normally distributed residual
res= np.random.normal(20, 10 ,100)
#calculates ground truth
y=10+5*x+res
#put data into dataframe for visualization and regression
df=pd.DataFrame({'x':x,'y':y})
df.head()
```

Linear Regression

```python
from matplotlib import pyplot as plt
import seaborn as sns
#put data from dataframe into a regression plot
sns.regplot(x='x', y='y', data=df)
plt.show()
```
