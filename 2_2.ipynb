2.2 Probability

Conditional Probability
Example Question:

A family has two children.

What is the probability that both are girls, given that the older child is a girl?

What is the probability that both are girls, given that at least one of the children is a girl?

```python
import random
#random kid 0 is female 1 is male
def random_kid():
  return random.randint(0,1)

both_girls=0
older_girl=0
one_girl=0

for x in range(10000):
  younger=random_kid()
  older=random_kid()
  if older == 0:
    older_girl+=1
  if older == 0 and younger ==0:
    both_girls+=1
  if older == 0 or younger == 0:
    one_girl+=1

print("Probability that both are girls given the older child is a girl: ", both_girls/older_girl)
print("Probability that both are girls given that at least one child is a girl: ",both_girls/one_girl)
```

Poisson Distribution

```python
# importing poisson from scipy
from scipy.stats import poisson

# importing numpy as np
import numpy as np

# importing matplotlib as plt
import matplotlib.pyplot as plt


# creating a numpy array for x-axis
x = np.arange(0, 100, 0.5)

# poisson distribution data for y-axis
y = poisson.pmf(x, mu=40, loc=10)


# plotting the graph
plt.plot(x, y)

# showing the graph
plt.show()
```
